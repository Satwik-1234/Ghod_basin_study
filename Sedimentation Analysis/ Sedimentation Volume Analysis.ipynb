```python
# Here is the Python code that performs the volume calculation using the Average End-Area Method, as broken down in the previous steps.

# This code defines the four measurement points, calculates the mean depth and cross-sectional area at each point, and then computes the total sediment volume assuming the points are equally spaced over a $15.00 \text{ m}$ total length.

import numpy as np

# --- 1. Input Data ---
# Structure dimensions
TOTAL_LENGTH = 15.00  # Total length of the sediment wedge along the stream (m)
NUM_SEGMENTS = 3      # Number of segments between 4 measurement points
L_SEG = TOTAL_LENGTH / NUM_SEGMENTS # Length of each segment (m)

# Data for 4 measurement points: [Width (m), Right Depth (m), Left Depth (m)]
data = {
    'Start': [4.00, 0.25, 0.20],
    'Middle': [2.30, 0.15, 0.12],
    'Second_last': [1.90, 0.05, 0.56],
    'Last': [1.60, 0.00, 0.00]
}

# --- 2. Calculate Cross-Sectional Area (A) at Each Point ---
# Formula: A = Width * ((D_R + D_L) / 2)
areas = {}
print("## Cross-Sectional Area Calculation")
print("------------------------------------")
for point, values in data.items():
    W, D_R, D_L = values

    # Calculate Mean Depth
    mean_depth = (D_R + D_L) / 2

    # Calculate Area
    area = W * mean_depth
    areas[point] = area

    print(f"Point: {point}")
    print(f"  Mean Depth: {mean_depth:.3f} m")
    print(f"  Area: {W:.2f} m * {mean_depth:.3f} m = {area:.4f} m^2\n")

# --- 3. Calculate Total Volume (V_Total) using Average End-Area Method ---
points = list(areas.keys())
total_volume = 0.0
segment_volumes = {}

print("## Segment Volume Calculation (L_seg = 5.00 m)")
print("-----------------------------------------------")
for i in range(NUM_SEGMENTS):
    point_name_1 = points[i]
    point_name_2 = points[i+1]

    A1 = areas[point_name_1]
    A2 = areas[point_name_2]

    # Formula: V_seg = ((A1 + A2) / 2) * L_seg
    V_seg = ((A1 + A2) / 2) * L_SEG
    total_volume += V_seg
    segment_volumes[f"{point_name_1} to {point_name_2}"] = V_seg

    print(f"Segment {i+1} ({point_name_1} to {point_name_2}):")
    print(f"  V = (({A1:.4f} + {A2:.4f}) / 2) * {L_SEG:.2f} = {V_seg:.3f} m^3")

# --- 4. Final Result ---
print("\n## Final Results")
print("-------------------")
print(f"Total Sediment Volume (V_Total): {total_volume:.3f} cubic meters (m^3)")
```

# Ghatghar Gabion Sedimentation

```python
# -----------------------------------------------
# SEDIMENT VOLUME CALCULATION (STEP-BY-STEP)
# Using Average Area Method
# Length = 25 m
# -----------------------------------------------

# 1) INPUT WIDTHS AND DEPTHS AT EACH SECTION
w0, d0 = 4.00, 0.350     # Start section
w1, d1 = 2.30, 0.250     # Middle section
w2, d2 = 1.90, 0.100     # Second-last section
w3, d3 = 1.60, 0.000     # Last section

# 2) COMPUTE CROSS-SECTIONAL AREAS
A0 = w0 * d0
A1 = w1 * d1
A2 = w2 * d2
A3 = w3 * d3

print("----- STEP 1: AREA CALCULATIONS -----")
print(f"A0 = {w0}  {d0} = {A0:.6f} m")
print(f"A1 = {w1}  {d1} = {A1:.6f} m")
print(f"A2 = {w2}  {d2} = {A2:.6f} m")
print(f"A3 = {w3}  {d3} = {A3:.6f} m")

# 3) SUM OF AREAS
sum_area = A0 + A1 + A2 + A3
print("\n----- STEP 2: SUM OF AREAS -----")
print(f"A0 + A1 + A2 + A3 = {sum_area:.6f} m")

# 4) AVERAGE CROSS-SECTIONAL AREA
Average_Area = sum_area / 4
print("\n----- STEP 3: AVERAGE AREA -----")
print(f"Average Area = {sum_area:.6f} / 4 = {Average_Area:.6f} m")

# 5) LENGTH OF SEDIMENT
L = 25  # meters
print("\n----- STEP 4: LENGTH USED -----")
print(f"Length L = {L} m")

# 6) FINAL VOLUME
Volume = Average_Area * L
print("\n----- STEP 5: FINAL VOLUME -----")
print(f"Volume = {Average_Area:.6f}  {L} = {Volume:.6f} m")

print("\n----------------------------------------")
print(f"FINAL SEDIMENT VOLUME = {Volume:.3f} m")
print("----------------------------------------")
```
